```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Load the CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalize pixel values
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Convert labels to categorical
num_classes = 10
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
# Define the CNN architecture
def create_model():
    model = keras.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model
```

```python
# Create models with and without data augmentation
model_augmented = create_model()
model_normal = create_model()
```

```python
# Compile the models
model_augmented.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model_normal.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Define data augmentation parameters
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.1
)
```

```python
# Fit the data generator on the training data
datagen.fit(x_train)
```

```python
# Train the models
epochs = 10
batch_size = 128
```

```python
print("Training model with data augmentation:")
model_augmented.fit(datagen.flow(x_train, y_train, batch_size=batch_size),
                    steps_per_epoch=len(x_train) // batch_size,
                    epochs=epochs,
                    validation_data=(x_test, y_test))

print("Training model without data augmentation:")
model_normal.fit(x_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(x_test, y_test))
```

```python
# Evaluate the model with data augmentation
print("Evaluating model with data augmentation:")
_, accuracy_augmented = model_augmented.evaluate(x_test, y_test, verbose=0)
print(f"Test accuracy with data augmentation: {accuracy_augmented:.4f}")
```

```python
# Evaluate the model without data augmentation
print("Evaluating model without data augmentation:")
_, accuracy_normal = model_normal.evaluate(x_test, y_test, verbose=0)
print(f"Test accuracy without data augmentation: {accuracy_normal:.4f}")
```
