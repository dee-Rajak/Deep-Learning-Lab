```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Perceptron:
    def __init__(self, input_size, learning_rate=0.01, epochs=100):
        self.weights = np.random.rand(input_size)
        self.bias = np.random.rand()
        self.learning_rate = learning_rate
        self.epochs = epochs

    def activation_function(self, x):
        return 1 if x >= 0 else 0

    def predict(self, X):
        weighted_sum = np.dot(X, self.weights) + self.bias
        return self.activation_function(weighted_sum)

    def train(self, X, y):
        for _ in range(self.epochs):
            for i in range(len(X)):
                prediction = self.predict(X[i])
                error = y[i] - prediction
                self.weights += self.learning_rate * error * X[i]
                self.bias += self.learning_rate * error
```

```python
# XOR gate inputs and labels
X_xor = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_xor = np.array([0, 1, 1, 0])
```

```python
# Create and manually adjust perceptron for XOR gate
perceptron_xor = Perceptron(input_size=2)
# Manually adjusting weights and bias for XOR gate
perceptron_xor.weights = np.array([1, 1])  # Adjusted weights
perceptron_xor.bias = -0.5  # Adjusted bias
```

```python
# Test perceptron for XOR gate
print("Testing XOR gate:")
for i in range(len(X_xor)):
    output = perceptron_xor.predict(X_xor[i])
    print(f"Input: {X_xor[i]}, Predicted Output: {output}")
```

```python
# Plot decision boundary before training
plt.figure(figsize=(6, 6))
plt.scatter(X_xor[:, 0], X_xor[:, 1], c=y_xor, cmap='coolwarm')
plt.title("Decision Boundary before Training")
plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.grid(True)
plt.show()
```

```python
# Generate linearly separable dataset
np.random.seed(0)
X_linear = np.random.randn(100, 2)
y_linear = np.random.randint(0, 2, size=100)
```

```python
# Create and train perceptron for linearly separable dataset
perceptron_linear = Perceptron(input_size=2)
perceptron_linear.train(X_linear, y_linear)
```

```python
# Plot decision boundary after training
plt.figure(figsize=(6, 6))
plt.scatter(X_linear[:, 0], X_linear[:, 1], c=y_linear, cmap='coolwarm')
plt.title("Decision Boundary after Training")
plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.grid(True)
```

```python
# Plot decision boundary
x_values = np.array([-2, 2])
y_values = -(perceptron_linear.weights[0] * x_values + perceptron_linear.bias) / perceptron_linear.weights[1]
plt.plot(x_values, y_values, color='black')
plt.show()
```
