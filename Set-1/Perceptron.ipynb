$y = 1, \mbox{if} \sum_i w_i x_i >= b$

$y =  0, \mbox{otherwise}$

```
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```
class Perceptron:

  # Initializes the parameters, values will be updated during training
  def __init__ (self):
    self.w = None
    self.b = None

  #model takes a single input 'x' and returns 1 if the weighted sum is greater than 'b'
  def model(self, x):
    return 1 if (np.dot(self.w, x) >= self.b) else 0

  #Takes set of input vectors 'X' and passes it to the model to make predictions and then stores the results in 'Y' (numpy array)
  def predict(self, X):
    Y = []
    for x in X:
      result = self.model(x)
      Y.append(result)
    return np.array(Y)



  #Takes training data X (features) and Y (labels), number of epochs epochs, and learning rate lr.
    #Initializes the weights self.w and bias self.b.
    #Contains a training loop that runs for a given number of epochs.
    #Within each epoch, it iteratively updates the weights and bias based on the perceptron learning rule:
        #If the actual label is 1 and the predicted label is 0, it increments the weights by lr * x and decrements the bias.
        #If the actual label is 0 and the predicted label is 1, it decrements the weights by lr * x and increments the bias.
    #The accuracy dictionary stores the accuracy of the model at each epoch.
    #The model keeps track of the highest accuracy and the corresponding weights and bias.
    #At the end of training, the weights and bias are set to the values that achieved the highest accuracy.
    #Plots the accuracy over epochs using Matplotlib.
    #Returns a matrix of weights from each epoch.

  def fit(self, X, Y, epochs = 1, lr = 1):

    self.w = np.ones(X.shape[1])
    self.b = 0

    accuracy = {}
    max_accuracy = 0

    wt_matrix = []

    for i in range(epochs):
      for x, y in zip(X, Y):
        y_pred = self.model(x)
        if y == 1 and y_pred == 0:
          self.w = self.w + lr * x
          self.b = self.b - lr * 1
        elif y == 0 and y_pred == 1:
          self.w = self.w - lr * x
          self.b = self.b + lr * 1

      wt_matrix.append(self.w)

      accuracy[i] = accuracy_score(self.predict(X), Y)
      if (accuracy[i] > max_accuracy):
        max_accuracy = accuracy[i]
        chkptw = self.w
        chkptb = self.b

    self.w = chkptw
    self.b = chkptb

    print(max_accuracy)

    plt.plot(accuracy.values())
    plt.ylim([0, 1])
    plt.show()

    return np.array(wt_matrix)
```

```
perceptron = Perceptron()
```

```
wt_matrix = perceptron.fit(X_train, Y_train, 10000, 0.5)
```

```
Y_pred_test = perceptron.predict(X_test)
print(accuracy_score(Y_pred_test, Y_test))
```

```
#The -1 index refers to the last item in the sequence, so this selects the final weights computed after all training epochs.
plt.plot(wt_matrix[-1,:])
plt.show()
```

```
%matplotlib inline
from matplotlib import animation, rc
from IPython.display import HTML
```

```
# First set up the figure, the axis, and the plot element we want to animate
fig, ax = plt.subplots()

ax.set_xlim(( 0, wt_matrix.shape[1]))
ax.set_ylim((-15000, 25000))

line, = ax.plot([], [], lw=2)
```

```
# animation function. This is called sequentially
def animate(i):
    x = list(range(wt_matrix.shape[1]))
    y = wt_matrix[i, :]
    line.set_data(x, y)
    return (line,)
```

```
# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, frames=100, interval=200, blit=True)
```
