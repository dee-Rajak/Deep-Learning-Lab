```python
import pandas as pd
import numpy as np
import random
```

```python
rand_data = []

for i in range(1000):
    rand_data.append([random.randint(25,50), random.randint(10, 250) *10000, random.randint(2,6), random.randint(0,1)])

df = pd.DataFrame(rand_data, columns=['Age', 'Salary', 'CountFamilyMembers', 'BuysFlat'])
display(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('BuysFlat', axis=1)
Y=df['BuysFlat']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.3, stratify=Y)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
X_binarised_train=X_train.apply(pd.cut , bins=2, labels=[1,0])
```

```python
X_binarised_test = X_test.apply(pd.cut, bins=2, labels=[1, 0])
X_binarised_test = X_binarised_test.values
X_binarised_train = X_binarised_train.values
```

```python
from sklearn.metrics import accuracy_score
```

```python
class MPNeuron:

    def __init__(self):
        self.b = None

    def model(self, x):
        return(sum(x) >= self.b)

    def predict(self, X):
        Y = []
        for x in X:
            result = self.model(x)
            Y.append(result)
        return np.asarray(Y)

    def fit(self, X, Y):
        accuracy = {}

        for b in range(X.shape[1] + 1):
            self.b = b
            Y_pred = self.predict(X)
            accuracy[b] = accuracy_score(Y_pred, Y)

        best_b = max(accuracy, key = accuracy.get)
        self.b = best_b
        for i in range(4):
            print('value of b is ', i)
            print('Accuracy is ', accuracy[i])
        print('value of Optimal b is ', best_b)
        print('Highest accuracy is ', accuracy[best_b])
```

```python
mp_neuron = MPNeuron()
mp_neuron.fit(X_binarised_train, Y_train)
```
